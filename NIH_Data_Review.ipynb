## NIH Data from multiple cameras
##### this is data shared from Dr. Evgeny Kobrinski's lab at the NIH, from their original camera and two replacement cameras.  Questions to a.wise@andor.com

````
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive
import os
drive.mount('/content/drive')
````

````
data_file = 'closedshutter1.fits'
data_dir = '/content/drive/My Drive/Andor/NIH MD Data'
````

````
os.listdir(data_dir)
````

````
for data_file in [f for f in os.listdir(data_dir) if '.fits' in f]:
  file_path = os.path.join(data_dir, data_file)
  data = fits.open(file_path)

  print('Filename is : ',data_file)
  for k in data[0].header:
    print(k,' = ',data[0].header[k],' , ',data[0].header.comments[k])
  camera_serial = 'X-' + str(data[0].header['SERNO'])
  print( 'Camera Serial - ',camera_serial)
  date_stamp = data[0].header['FRAME']
  print('Date / Time Stamp - ' + date_stamp)

  def meanmed(x, axis):
    return np.mean(x, axis) - np.median(x, axis)

  functions = {'Median':np.median, 'Mean':np.mean, 'STD':np.std}

  fig, axarr = plt.subplots(nrows = 1, ncols = len(functions), figsize = (14,6))

  for i, (name, f) in enumerate(functions.items()):
    result = f(data[0].data, axis=0)
    lims = np.quantile(result.ravel(),[0.05,0.95])
    im = axarr[i].imshow(result, vmin = lims[0], vmax = lims[1])
    axarr[i].set_title(camera_serial + ' - Per Pixel ' + name + '\n' + date_stamp.replace('T',' ') )
    fig.colorbar(im, ax = axarr[i], fraction=0.046, pad=0.04)
    axarr[i].set_xticks([])
    axarr[i].set_yticks([])
  plt.savefig(data_file + '_summary_images' + '.png')
  plt.subplots_adjust(wspace=0.4)
  plt.show()

  lims_adjust = {'Median':[-2,2], 'Mean':[-2,2], 'STD':[-.1,.1], 'Mean-Med': [0,0]}
  fig, axarr = plt.subplots(ncols = 1, nrows = len(functions), figsize = (14,9))

  for i, (name, f) in enumerate(functions.items()):
    time_series = f(data[0].data.reshape(data[0].shape[0],-1), axis = 1)
    lims = np.quantile(time_series,[0.01,0.98])
    axarr[i].set_title(camera_serial + ' - Per Exposure ' + name + ' - ' + date_stamp.replace('T',' ') )
    axarr[i].plot(time_series,'k-', alpha = 0.5)
    axarr[i].grid(True)
    axarr[i].set_ylim([ lims[0] + lims_adjust[name][0], lims[1] + lims_adjust[name][1]] )
    axarr[i].set_ylabel('Counts -' + name)
    axarr[i].set_xlabel('Exposure #')
  plt.savefig(data_file + '_time_series' + '.png')
  plt.subplots_adjust(hspace=0.5)
  plt.show()
````
